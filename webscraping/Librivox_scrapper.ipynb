# Web scraping script #2: fetching synopsis from LibriVox

```python
from bs4 import BeautifulSoup
import requests
import pickle
from re import search
from time import sleep
import pandas as pd
import sys
path='../'
```

```python
def get_librivox_df():
    lst=[]
    for i in range(50,15500):
        weblink='https://librivox.org/api/feed/audiobooks/?id='+str(i)+'&format=json'
        page=requests.get(weblink)
        try:
            lst.append([page.json()['books'][0]['title'],
                page.json()['books'][0]['description'],
                page.json()['books'][0]['language'],
                page.json()['books'][0]['url_librivox'],
                page.json()['books'][0]['totaltime'],
                page.json()['books'][0]['authors'][0]['first_name'] + ' '+ page.json()['books'][0]['authors'][0]['last_name']])
        except:
            pass
    final_df=pd.DataFrame(lst,columns=['Book Name','Book Synopsis','Language','URL','Time','Authors'])
    final_df.to_csv('librivox_synopsis_df.csv')

```

```python
get_librivox_df()
```

```python
def fetch_and_clean_data(fname='librivox_synopsis_df.csv'):
    df=pd.read_csv(fname)
    cleaned=df.drop('Unnamed: 0',1)
    cleaned=cleaned[cleaned['Language']=="English"]
    return cleaned
```

```python
df=fetch_and_clean_data()
mydict = dict(zip(df['Book Name'], df['URL']))
Genre=dict()
for key,val in mydict.items():
    try:
        syn_page=requests.get(val)
        syn_page_details= BeautifulSoup(syn_page.content, 'html.parser')
        Genre[key]=syn_page_details.findAll('p',{'class':'book-page-genre'})[0].get_text()
    except:
        Genre[key]='None'
        pass
```

```python
df['Genre']=df['Book Name'].apply(lambda x:Genre[x].split(':')[-1])
```

```python
df.to_csv('librivox_synopsis_df_genre.csv')
```
